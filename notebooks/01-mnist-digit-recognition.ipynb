```python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)

print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
plt.figure(figsize= (10,4))
for i in range(10):
    plt.subplot(2,5, i+1) #2 satır, 5 sütun, i+1. hücre
    plt.imshow(X_train[i], cmap='gray') #imshow -> görüntüyü çiz, cmap='gray' -> siyah beyaz
    plt.title(f"Label: {y_train[i]}") #bu görüntünün gerçek rakamı
    plt.axis('off')
plt.show()
```

```python
import numpy as np
unique, counts = np.unique(y_train, return_counts=True) #fonksiyonu, y_train dizisindeki tekrar eden değerleri temizleyerek sadece benzersiz olanları bulur
# return_counts=True parametresi, bu benzersiz değerlerin her birinin veri setinde kaç kez geçtiğini hesaplar ve counts adlı bir dizi olarak döndürür

for label, count in zip(unique, counts):
    print(f"Label {label}: {count} samples")
```
